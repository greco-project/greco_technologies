```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/home/adminlocal/Schreibtisch/Präsi_Modeling_PeroSI/redigitalize/EQE_3T_peroskit_measured.csv", sep=";", index_col=0, header=None, names=['wavelength', 'EQE'])
df=df.sort_index()
df.index=df.index*10
df.plot()
df
```

```python
new_index= [*range(2800, 8000, 10)]
```

```python
df_out = pd.DataFrame(index=new_index)
df_out.index.name = df.index.name

for colname, col in df.iteritems():
        df_out[colname] = np.interp(new_index, df.index, col)

df_out.index=df_out.index/10
df_out.plot()
```

```python
df_out=df_out.rename(columns={0: "wavelength", 1: "EQE"})
```
